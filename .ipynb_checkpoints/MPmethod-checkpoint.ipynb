```python
# import os
# import pandas as pd
# from collections import defaultdict
from docplex.mp.model import Model
from docplex.mp.progress import ProgressDataRecorder
import matplotlib.pyplot as plt
key_shift=	0
melody_input=[ 40, 40, 41, 43, 43, 41, 40, 38, 36, 36, 38, 40, 38, 36, 36, 36]
melody_input=[36, 36, 38, 40, 38, 36, 36, 36]
#melody_input=[a-16 for a in melody_input]

tonality =1 # "major" #or 0 "minor"
N =len(melody_input)

chord_list0=[[0, 3, 7], # i
            [0, 4, 7], # I
            [2, 5, 9], # ii
            [2, 6, 9], # II
            [4, 7, 11], # iii
            [4, 8, 11], # III
            [5, 8, 0], # iv
            [5, 9, 0], # IV
            [7, 10, 2], # v
            [7, 11, 2], # V
            [9, 0, 4], # vi
            [9, 1, 4], # VI
            [11, 2, 5], # vii_dim
            [11, 2, 6], # vii
            [11, 3, 6]] # VII
chord_list=[[0, 3, 7], # i
            [0, 4, 7], # I
            [2, 5, 9], # ii
            [2, 6, 9], # II
            [4, 7, 11], # iii
            [4, 8, 11], # III
            [5, 8, 0], # iv
            [5, 9, 0], # IV
            [7, 10, 2], # v
            [7, 11, 2], # V
            [9, 0, 4], # vi
            [9, 1, 4], # VI
            [11, 2, 5], # vii_dim
            [11, 2, 6], # vii
            [11, 3, 6]] # VII

no_chord=len(chord_list)
for k in range(1,4):
    for l in range(no_chord):
        chord_list[l]+=([x+12*k for x in chord_list0[l]])
        chord_list[l]+=([x-12*k for x in chord_list0[l]])
    

m = Model(name="Harmony")


# x[i,j]	Notes
# j belongs to {0, ..., N-1}, where n is the number of chords in the sequence
# i represents voices (1=alto, 2=tenor, 3=bass). 
# x[0,j] refers to the soprano voice, but this is provided and not a decision variable

arr=[(i,j) for i in range(4) for j in range(N)]
x= m.integer_var_dict(arr, 0, 50, "notes")  # !!!todo:  fix lower/upper bound by Key

# c[j]	Chords
c = m.integer_var_list(N, 0, no_chord-1, "chords")

# =============================================================================
# Hard constraints   
lb, ub=[19,12,5],[38,28,26]
for j  in range(N):  
    for i in range(1,4):
        m.add_constraint(x[i,j]>=lb[i-1]) 
        m.add_constraint(x[i,j]<=ub[i-1])
         

##notes must be in the same chord
for j in range(N):
    m.add_constraint(x[0,j]==melody_input[j] ) # read input
    for i in range(0,4):
        for l in range(no_chord):
            m.add_constraint(  (c[j]==l)<= m.sum  ((x[i,j]-24) == chord_list[l][p] for p in range(21))     )
            
            
# #Tonality Requirements -first/last grade
#m.add_constraint((tonality==1)<=(c[0]==0))
#m.add_constraint((tonality==0)<=(c[0]==1))
m.add_constraint(tonality==c[0])
m.add_constraint(c[N-1] <=1 )
m.add_constraint(tonality<=c[N-1])



for j in range(N-1):
    if j%4==3:
        m.add_constraint(c[j]!=c[j+1])
           
# #forbidden to write voice overlap (strictly between soprano, alto and tenor, not strictly between tenor and bass):
for j in range(N):
    m.add_constraint(x[0,j]>=x[1,j]+1)
    m.add_constraint(x[1,j]>=x[2,j]+1)
    m.add_constraint(x[2,j]>=x[3,j])
    
# for j in range(N-1):  # !!!change to soft constraint?
#     m.add_constraint(x[0,j]>x[1,j+1])
#     m.add_constraint(x[1,j]>x[2,j+1])
#     m.add_constraint(x[2,j]>=x[3,j+1])


#  # No parallel fifths or octaves are allowed:  
for j in range(N-1):
    for i in range(3):
        for k in range(i+1,4):
            m.add_constraint((x[i,j]-x[k,j]==7)<=(x[i,j+1]-x[k,j+1]!=7))
            m.add_constraint((x[i,j]-x[k,j]==19)<=(x[i,j+1]-x[k,j+1]!=19))
            m.add_constraint((x[i,j]-x[k,j]==31)<=(x[i,j+1]-x[k,j+1]!=31))
            m.add_constraint((x[i,j]-x[k,j]==12)<=(x[i,j+1]-x[k,j+1]!=12))
            m.add_constraint((x[i,j]-x[k,j]==24)<=(x[i,j+1]-x[k,j+1]!=24))
            m.add_constraint((x[i,j]-x[k,j]==36)<=(x[i,j+1]-x[k,j+1]!=36))

# =============================================================================            
            
            
# =============================================================================
# Soft constraints          

#Grade transition constraints 
dic_chord_progression={(2,5) : 0,(5,1) : 0,(1,2) : 0,(4,5) : 0,(6,5) : 0,(6,2) : 0,(1,4) : 0,(5,6) : 0,(6,4) : 0,(5,1) : 0,(1,6) : 0,(1,1) : 0.3,(5,5) : 0.3,(6,1) : 0.3,(6,6) : 0.3,(4,2) : 0.4,(4,4) : 0.4,(2,2) : 0.4,(1,5) : 0.5,(4,1) : 1,(2,1) : 1,(2,6) : 1}      
def cost_chord_progression(c1,c2):
    return(  dic_chord_progression.get((chord_list[c1][0],chord_list[c2][0]),1000) )
    
# #cost= sum([cost_chord_progression(c[l],c[l+1]) for l in range(N-1)]) 
cost1= m.continuous_var_list(N, 0,100000, "Progression cost")            
for j in range(N-1):
    for c1 in range(no_chord):
        for c2 in range(no_chord):
            m.add_constraint(m.logical_and(c1==c[j],c2==c[j+1])*cost_chord_progression(c1,c2)<=  cost1[j]) 
    
#m.add_indicator_constraint(m.indicator_constraint(m.less(x[i,j],slb[i-1]),cost3[j]==chg*(slb[i-1]-x[i,j])+stepl[i-1])) 
#m.add_indicator_constraint(m.indicator_constraint(m.greater(x[i,j],sub[i-1]),cost3[j]==chg*(x[i,j]-sub[i-1])+stepu[i-1])) 
   
chg=0.11   
slb, sub,stepl, stepu=[24,17, 10 ],[33,23 ,20 ], [0,0.1,0.1]  ,[0.4,0.3,0.1]      
cost3= m.continuous_var_list(N, 0,40, "Voice range cost")                 
for j  in range(N):  
    for i in range(1,4):
        m.add_constraint(cost3[j]>=chg*(slb[i-1]-x[i,j])+stepl[i-1])
        m.add_constraint(cost3[j]>=chg*(x[i,j]-sub[i-1])+stepu[i-1])


    

    #Contrary motion constraints:
    #there is no contrary motion between bass and soprano.    
cost5= m.binary_var_list(N, 0,40, "Contrary motion cost")

for j  in range(N-1): 
    m.add_constraint((m.logical_and(x[0,j]<=x[0,j+1]-1,x[3,j]>=x[3,j+1]-1)<=cost5[j]))
    m.add_constraint((m.logical_and(x[0,j]>=x[0,j+1]+1,x[3,j]<=x[3,j+1]+1)<=cost5[j] ))
    #m.add_constraint((m.logical_and(x[0,j]==x[0,j+1],x[3,j]==x[3,j+1])<=cost5[j]))
    #m.add_constraint((m.logical_and(x[0,j]!=x[0,j+1],x[3,j]==x[3,j+1])<=cost5[j]))
    
#m.minimize(m.sum(cost3[j]+cost1[j]+cost5[j] for j in range(N)))        
m.minimize(m.sum(cost1[j]+cost3[j]+cost5[j] for j in range(N)))          
#sol = m.solve(log_output = True)
```

```python
m.context.update_cplex_parameters({'randomseed': 606, 'mip.tolerances.mipgap': 0.02,'timelimit': 60})
```

```python
recorder = ProgressDataRecorder()
m.add_progress_listener(recorder)
sol=m.solve()   
print(sol)
```

```python
best_bound = []
det_time = []
current_objective = []
for data in recorder.recorded:
    best_bound.append(data.best_bound)
    det_time.append(data.det_time/100)
    current_objective.append(data.current_objective)
```

```python
plt.plot(det_time, current_objective, label = "best integer", marker='s', linestyle = '--')
plt.plot(det_time, best_bound, label = "best node")
plt.legend()
plt.show()
```
